1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
df_raw = pd.read_csv('previsao_de_renda.csv')
df_raw.head(3)
````

````python
df = df_raw.drop('Unnamed: 0', axis=1)
````

````python
df.info()
````

````python
df = df_raw.dropna()
df.isna().sum()
````

````python
# Estilo do gráfico
sns.set(style="whitegrid")
````

````python
# Gráfico 1
plt.figure(figsize=(10, 6))
ax1 = sns.pointplot(x='posse_de_veiculo', y='renda', data=df, dodge=True, errorbar=('ci', 95))
plt.title('Renda Média por Posse de Veículo')
plt.ylabel('Renda Média')
plt.xlabel('Posse de Veículo')
plt.show()
````

````python
# Gráfico 2

plt.figure(figsize=(10, 6))
ax2 = sns.pointplot(x='posse_de_imovel', y='renda', data=df, dodge=True, errorbar=('ci', 95))
plt.title('Renda Média por Posse de Imóvel')
plt.ylabel('Renda Média')
plt.xlabel('Posse de Imóvel')
plt.show()
````

## Interpretação:

### Ambos os gráficos apresentam(_posse_ _de_ _imóvel_ e variável _posse_ _de_ _veículo_) apresentam uma tendência positiva e indicam serem boas preditoras de um cliente com maior renda.
